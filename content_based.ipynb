```python
items = items.groupby('asin', as_index=False).mean()
```

```python
items
```

```python
def prediction(s):
    user_index = user.index[user['reviewerID'] == s]
    items_consumed = df1_new['asin'][df1_new['reviewerID'] == s]
    items_consumed = list(items_consumed)
    cosine = np.dot(user.iloc[user_index,1:],(items.iloc[:,1:] ).T)
    suggest = (pd.DataFrame(cosine))
    suggest.columns = items['asin']
    a = suggest.columns
    suggest = suggest.T
    suggest['item'] = a
    suggest = suggest.iloc[:-1,:]
    suggest = suggest.sort_values(by=[0], ascending=False)
    item_list = suggest['item']
    item_suggest =  [x for x in item_list if x not in items_consumed]
    return item_suggest[:10]
```
